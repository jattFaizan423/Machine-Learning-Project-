```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/quikr_car.csv')
```

```python
df.head()
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.shape
```

# **1.Data Cleaning**

```python
df['name'] = df['name'].str.split().str.slice(0,3).str.join(' ')
```

```python
df['name'] = df['name'].astype('category')
```

```python
df['company'] = df['company'].astype('category')
```

```python
df = df[df['year'].str.isnumeric()].copy()
```

```python
df['year'] = df['year'].astype('int32')
```

```python
df = df[df['Price'] != 'Ask For Price']
```

```python
df['Price'] = df['Price'].str.replace(',','').astype(float)
```

```python
df['kms_driven'] = df['kms_driven'].str.split().str[0].str.replace(',','')
```

```python
df = df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df['fuel_type'] = df['fuel_type'].fillna('Petrol').astype('category')
```

```python
df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df = df[df['Price'] < 6e6].reset_index(drop=True)
```

```python
df = df[df['kms_driven'] < 200000].reset_index(drop=True)
```

```python
df.head()
```

# **2.Apply Model**

```python
X = df.drop(columns='Price')
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=478)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
columns_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(columns_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
