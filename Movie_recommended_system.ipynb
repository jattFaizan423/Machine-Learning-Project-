```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv', on_bad_lines='skip', engine='python') # Trying 'python' engine for potentially more robust parsing
credits = pd.read_csv('/content/tmdb_5000_credits.csv', on_bad_lines='skip', engine='python') # Trying 'python' engine for potentially more robust parsing
```

```python
movies.head(2)
```

```python
credits = credits[[	'movie_id','title','cast','crew']]
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
import ast
def convert(text):
  l = []
  for i in ast.literal_eval(text):
    l.append(i['name'])
  return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
import json
def fetch_director(text):
    L = []
    if isinstance(text, str):
        try:
            for i in json.loads(text):
                if i['job'] == 'Director':
                    L.append(i['name'])
        except json.JSONDecodeError:
            pass # Handle cases where the string is not valid JSON
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=7000, stop_words='english')
```

```python
vector = tfidf.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```

```python
import pickle
pickle.dump(df,open('movie_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
